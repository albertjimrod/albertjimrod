### Porque aparece `SettingWithCopyWarning`???

El problema reside en que la asignación de la **indexación encadenada** tiene resultados __inherentemente impredecibles__. 

Para ver esto, piense en cómo el intérprete de Python ejecuta este código:

    dfmi.loc[:, ('one', 'second')] = value
    # becomes
    dfmi.loc.__setitem__((slice(None), ('one', 'second')), value)
    
Pero este código se maneja de manera diferente:

        dfmi['one']['second'] = value
        # becomes
        dfmi.__getitem__('one').__setitem__('second', value)
        
¿ Ves **__getitem__** ahí? 

Fuera de los casos simples, es muy difícil predecir si devolverá una vista o una copia (depende del diseño de la memoria de la matriz, sobre el cual pandas no ofrece garantías) y, por lo tanto, si el `__setitem__` modificará **dfmi** o un **objeto temporal** que se desecha inmediatamente después. ¡Eso es lo que `SettingWithCopy` le está advirtiendo!

A veces `SettingWithCopy` aparecerá con una advertencia cuando no haya una indexación encadenada obvia.

Estos son los errores que  `SettingWithCopy`está diseñado para atrapar! pandas probablemente esté tratando de advertirte que has hecho esto:

    def do_something(df):
        foo = df[['bar', 'baz']]  # Is foo a view? A copy? Nobody knows!
        # ... many lines here ...
        # We don't know whether this will modify df or not!
        foo['quux'] = value
        return foo
        
### El orden de evaluación importa

Cuando utiliza la indexación encadenada, el orden y el tipo de la operación de indexación determinan parcialmente si el resultado es una división del objeto original o una copia de la división.

pandas tiene el `SettingWithCopyWarning` porque la asignación a una copia de un segmento con frecuencia no es intencional, sino un error causado por la indexación encadenada que devuelve una copia donde se esperaba un segmento.

Si desea que los pandas confíe más o menos en la asignación a una expresión de indexación encadenada, puede establecer la opción `mode.chained_assignment` en uno de estos valores:

- `'warn'`: imprime un valor predeterminado, significa `SettingWithCopyWarning`.

- `'raise'`: significa que pandas generará un problema con el `SettingWithCopyException` que tendrás que lidiar.

- `None`: suprimirá las advertencias por completo.

         
        dfb = pd.DataFrame({'a': ['one', 'one', 'two',
                                      'three', 'two', 'one', 'six'],
                                'c': np.arange(7)})

         # This will show the SettingWithCopyWarning
         # but the frame values will be set
         dfb['c'][dfb['a'].str.startswith('o')] = 42

#### Sin embargo, esto está funcionando en una copia y no funcionará.

    pd.set_option('mode.chained_assignment','warn')
    dfb[dfb['a'].str.startswith('o')]['c'] = 42
    Traceback (most recent call last)
         ...
    SettingWithCopyWarning:
         A value is trying to be set on a copy of a slice from a DataFrame.
         Try using .loc[row_index,col_indexer] = value instead
         
#### Una asignación encadenada también puede surgir en la configuración de un marco de tipo mixto.


El siguiente es el método de **acceso recomendado**:

**`.loc` para múltiples elementos** (usando mask) y un solo elemento usando un índice fijo:

                              'three', 'two', 'one', 'six'],
                        'c': np.arange(7)})


    dfd = dfc.copy()

    # Setting multiple items using a mask
    mask = dfd['a'].str.startswith('o')

    dfd.loc[mask, 'c'] = 42

### 3 - Now that we're finished cleaning the data, we'll generate a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('default')

plt.plot(euro_dollar['Time'],euro_dollar['US_dollar'])
plt.show()
```

We see many small wiggles — rather than seeing a smooth line. The wiggles, however, have meaning: : they are the visual representation of the daily variation in the exchange rate. 
        
Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average).

Calculate the rolling means for the `US_dollar` column using a moving window of 30 days. Add the rolling means to a new column named `rolling_mean`.

```python
euro_dollar['rolling_mean'] = euro_dollar['US_dollar'].rolling(30).mean().copy()
```

```python
%matplotlib inline
plt.plot(euro_dollar['Time'],euro_dollar['rolling_mean'])
plt.show()
```

```python
%matplotlib inline
#style.use('Solarize_Light2')

fig, ax = plt.subplots()

ax.bar(euro_dollar['Time'],euro_dollar['rolling_mean'])

plt.show()
```

```python

ax.text(x=10000, y=1.8, s= 'Ratio Euro / Dollar',size=15, weight='bold')

fig, ax = plt.subplots()


ax.plot(euro_dollar['Time'],
       euro_dollar['rolling_mean'], color='#7cd6f0')

for location in ['left','right','bottom','top']:
    ax.spines[location].set_visible(False)
    
# Move ticks labels at the top of the graph
ax.xaxis.tick_top()

# Remove ticks on top and left of the graph
ax.tick_params(top = False, left = False)

# We'll x-tick labels in grey and bars in a shade of red.

ax.tick_params(axis='x', colors='grey')

plt.show()
```
